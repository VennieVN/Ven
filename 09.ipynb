```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# path img
my_path = 'C:/Users/Admin/Desktop/img/Anime5.jpg'
# read img
im = Image.open(my_path)
plt.imshow(im)
```

```python
# change img to Gray
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# Draw map of Histogram img
plt.figure()
plt.hist(im_array.flatten(), bins=150)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
# contour img
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
